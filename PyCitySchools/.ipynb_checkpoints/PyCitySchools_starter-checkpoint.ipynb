```python
# Dependencies and Setup
import pandas as pd

# Files to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset, have merged, but haven't used it yet - add a column later, need to move this
merged_data_df = pd.merge(school_data_df, student_data_df, how="left", on=["school_name", "school_name"])
```

```python
school_count = school_data_df['school_name'].count()
school_count
```

```python
student_total = student_data_df['student_name'].count()
student_total
```

```python
total_budget = school_data_df['budget'].sum()
total_budget
```

```python
ave_math = student_data_df['math_score'].mean()
ave_math
```

```python
ave_read = student_data_df['reading_score'].mean()
ave_read
```

```python
overall_pass_rate = (ave_math + ave_read)/2
overall_pass_rate
```

```python
# calc % of students with a passing math score 70+
passing_math = student_data_df.loc[student_data_df["math_score"]>=70]
students_passing_math = passing_math['math_score'].count()
math_pass_prcnt = (students_passing_math/student_total)*100
math_pass_prcnt
```

```python
# calc % of students with a passing math score 70+
passing_reading = student_data_df.loc[student_data_df["reading_score"]>=70]
students_passing_read = passing_reading['reading_score'].count()
read_pass_prcnt = (students_passing_read/student_total)*100
read_pass_prcnt
```

## My District Summary

```python
# Place all of the data found into a summary DataFrame
district_sum_results_data = [{"Total Schools": school_count, "Total Students": student_total, "Total Budget": total_budget,
                              "Average Math Score": ave_math, "Average Reading Score": ave_read,
                              "% Passing Math": math_pass_prcnt, "% Passing Reading": read_pass_prcnt,
                             "% Overall Passing Rate": overall_pass_rate}]

dist_results_data_df = pd.DataFrame(district_sum_results_data)

dist_results_data_df["Total Budget"] = dist_results_data_df["Total Budget"].map("${:.2f}".format)
dist_results_data_df["Total Students"] = dist_results_data_df["Total Students"].map("{:,}".format)

dist_results_data_df
```
